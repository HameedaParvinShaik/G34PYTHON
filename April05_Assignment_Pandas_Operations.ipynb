# <b>Problem 5 : Print the company name which manufactured the most expensive car. Also print the price of the most expensive car

```python
# Your code here

max_price_row = df.loc[df['price'].idxmax()]

print("Company:", max_price_row['make'])
print("Price:", max_price_row['price'])
```

# <b>Problem 6 : Calculate maximum horsepowers for each company. </b>

```python
# Your code here

max_hp_by_make = df.groupby('make')['horsepower'].max()

print(max_hp_by_make)
```

# <b> Problem 7 : Print the total number of cars manufactured by each company

```python
# Your code here
car_count_by_company = df['make'].value_counts()
print(car_count_by_company)
```

# <b> Problem 8 : Create a new column for the updated price. This updated price will be calculated on top of the given price - if the engine is in front, price will be same else if the engine is in rear, price will be doubled </b>


```python
# Your code here. Use apply and lambda

df['updated-price'] = df.apply(
    lambda row: row['price'] * 2 if row['engine-location'] == 'rear' else row['price'],
    axis=1
)
print(df.head(10))
```

# <b>Problem 9 : Sort the dataframe by car and price columns

```python
# Your code here
df_sorted = df.sort_values(by=['make', 'price'])

print(df_sorted[['make', 'price']].head(10))
```

# <b> Problem 10 : Create a new column which stores the number of doors in a car as integers.

```python
# Your code here. Use apply
df['num-of-doors-int'] = df['num-of-doors'].apply(
    lambda x: 2 if x == 'two' else 
              3 if x == 'three' else 
              4 if x == 'four' else 
              5 if x == 'five' else 
              6 if x == 'six' else np.nan
)

print(df[['num-of-doors', 'num-of-doors-int']].head(10))

```

# <b>Problem 11 : Concatenate the two data frames given below firstly row wise and secondly column wise.</b>

<code>GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}</code>

<code>japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}</code>

```python
# Your code here

GermanCars = {
    'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'],
    'Price': [23845, 171995, 135925, 71400]
}

japaneseCars = {
    'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '],
    'Price': [29995, 23600, 61500, 58900]
}

df_german = pd.DataFrame(GermanCars)
df_japanese = pd.DataFrame(japaneseCars)
```

```python
row_wise = pd.concat([df_german, df_japanese], ignore_index=True)
print("Row-wise concatenation:\n", row_wise)
```

```python
column_wise = pd.concat([df_german, df_japanese], axis=1)
print("\nColumn-wise concatenation:\n", column_wise)
```
